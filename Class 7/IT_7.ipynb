```python
# Linear Regression
import seaborn as sns
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression

# Load the dataset
df = sns.load_dataset('tips')
print(df)
x = df['total_bill'].values.reshape(-1, 1)
y = df['tip'].values
# print(x)
# print(y)

# Build relation between x and y
model = LinearRegression()
model.fit(x, y)

# Predict values
residuals = model.predict(x) - y
sm.qqplot(residuals, line='s')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 3, 5, 6, 8])

# Fit model
model = LinearRegression()
model.fit(x.reshape(-1, 1), y)

# plot
plt.scatter(x, y, color='blue', label='Data')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression')
plt.plot(x, model.predict(x.reshape(-1, 1)), color='red', label='Regression Line')
plt.legend()
plt.show()
```

```python
# Build the Linear Regression Model to predict housing prices
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Define our dataset
square_feet = np.array([1000, 1500, 2000, 2500, 3000])
house_prices = np.array([200000, 250000, 300000, 350000, 400000])

# Build the linear regression model
model = LinearRegression()  # Initialize the model
model.fit(square_feet.reshape(-1, 1), house_prices) # Train the model

# Make a prediction
predicted_price = model.predict(np.array([2000]).reshape(-1, 1))
print(f"Predicted price for a house with 2000 square feet: ${predicted_price[0]:.2f}")
```

```python
plt.scatter(square_feet, house_prices, color='blue', label='Data')
plt.plot(square_feet, model.predict(square_feet.reshape(-1, 1)), color='red', label='Regression Line')
plt.legend()
plt.xlabel('Square Feet')
plt.ylabel('House Prices')
plt.title('Linear Regression')
plt.legend()
plt.show()
```

```python
# Task 1
import numpy as np
import matplotlib.pyplot as plt
def linear_function(x, c=3, m=2):
  return (m*x + (c))

# Task 2
x = np.linspace(-10, 10, 100)
y = linear_function(x)

# Task 3
plt.plot(x, y, label="y = 2x + 3", color='#01a89e', linestyle='-', linewidth=2)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Linear Function")
plt.grid(True)
plt.legend()
plt.show()
```
